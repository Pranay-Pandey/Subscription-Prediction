#THE Brief implementation of the MODEL

```python
from xgboost.sklearn import XGBModel
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
import xgboost as xgb
import lightgbm
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

def model(X,Y):
  
  X = X.append(X)
  Y = Y.append(Y)
  X = X.append(X)
  Y = Y.append(Y)

  X['job'] = le.fit_transform(X['job'])
  X['marital'] = le.fit_transform(X['marital'])
  X['education'] = le.fit_transform(X['education'])
  X['default'] = le.fit_transform(X['default'])
  X['housing'] = le.fit_transform(X['housing'])
  X['loan'] = le.fit_transform(X['loan'])
  X['month'] = le.fit_transform(X['month'])
  X['poutcome'] = le.fit_transform(X['poutcome'])
  Y['Category'] = le.fit_transform(Y['Category'])

  X.sort_values(X.columns[0])
  Y.sort_values(Y.columns[0])

  X = X.drop([X.columns[0],'contact'], axis = 1)
  Y = Y.drop([Y.columns[0]], axis = 1)

  train_X = lightgbm.Dataset(X, label=Y)

  parameters = {'objective':'binary',
              'metric': 'accuracy',
              'is_unbalance': 'true',
              'boosting': 'gbdt',
              'num_leaves': 63,
              'feature_fraction': 0.5,
              'bagging_fraction': 0.5,
              'bagging_freq': 20,
              'learning_rate': 0.01,
              'verbose': -1
              }
  model_lgbm = lightgbm.train(parameters,
                            train_X,
                            num_boost_round = 5000)   

  X['Current'] = Y['Category']

  xgbm = xgb.XGBClassifier()
  xgbm.fit(X, Y)

  return model_lgbm,xgbm
```

```python
training_x = pd.read_csv('/content/xtrain_csv.csv')
training_y = pd.read_csv('/content/ytrain_csv.csv')
half_model,stackmodel = model(training_x,training_y)
```

#PREDICT

```python
def predict(datasetinput):
  from sklearn.preprocessing import LabelEncoder
  le = LabelEncoder()



  id_values = datasetinput[datasetinput.columns[0]].values

  datasetinput['job'] = le.fit_transform(datasetinput['job'])
  datasetinput['marital'] = le.fit_transform(datasetinput['marital'])
  datasetinput['education'] = le.fit_transform(datasetinput['education'])
  datasetinput['default'] = le.fit_transform(datasetinput['default'])
  datasetinput['housing'] = le.fit_transform(datasetinput['housing'])
  datasetinput['loan'] = le.fit_transform(datasetinput['loan'])
  datasetinput['month'] = le.fit_transform(datasetinput['month'])
  datasetinput['poutcome'] = le.fit_transform(datasetinput['poutcome'])
  datasetinput.sort_values(datasetinput.columns[0])
  datasetinput = datasetinput.drop([datasetinput.columns[0],'contact'], axis = 1)

  #LIGHTGBM prediction
  y_train_pred = half_model.predict(datasetinput)

  datasetinput['Current'] = np.where(y_train_pred>0.5,1,0)
  predictions = stackmodel.predict(datasetinput)

  finalprediction = np.where(predictions>0.5,"yes","no")

  data = {'Id':id_values,
          'Category':finalprediction}

  df = pd.DataFrame(data)
  
  return df
  
```

```python
test_file = pd.read_csv('/content/xtest_csv.csv')
file = predict(test_file)
```

#Testing On a Blind Dataset and calculating accuracy

A different test dataset can be downloded from https://drive.google.com/file/d/1n4sD8g_0K8tXFhmeBgu1_atL1b-1DbnT/view?usp=sharing to check the accuracy of the model

```python
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

def checkaccuracy(X_testing,Y_testing):
  prediction_Y = predict(X_testing)
  prediction_Y[prediction_Y.columns[1]] = le.fit_transform(prediction_Y[prediction_Y.columns[1]])
  Y_testing[Y_testing.columns[1]] = le.fit_transform(Y_testing[Y_testing.columns[1]])

  score = accuracy_score(Y_testing[Y_testing.columns[1]],prediction_Y[prediction_Y.columns[1]])

  return score
```

```python
blind_test_X = pd.read_csv('/content/Xtest_Feature.csv')
blind_test_Y = pd.read_csv('/content/Ytest_target.csv')
print(checkaccuracy(blind_test_X,blind_test_Y))
```
